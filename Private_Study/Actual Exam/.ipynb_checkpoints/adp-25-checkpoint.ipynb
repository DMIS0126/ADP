## 1.1 F(소비자별 구매빈도), M(소비자별 총 구매액) feature를 새로 생성해서 그 결과값으로 탐색적 분석 실시


```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/adp-p7/problem1.csv')

df.head()
```

```python
df.info()
```

```python
# Customer 수 확인
df['CustomerID'].unique().shape
```

```python
# 결측치가 없으므로 바로 F, M feature 생성

# InvoiceDate에서 시간 제거하고 일자만 남기기
df['date'] = pd.to_datetime(df['InvoiceDate']).dt.date
F = df.groupby('CustomerID').count()['date']

df['price'] = df['UnitPrice'] * df['Quantity']
M = df.groupby('CustomerID').sum('price')
M = M['price']

F.to_frame()
M.to_frame()

df_new = pd.concat([F, M], axis = 1)
df_new.columns = ['F', 'M']
```

```python
df_new.info()
```

```python
df_new.describe()
```

```python
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.hist(df_new['F'])
plt.xlabel('F')
plt.ylabel('cnt_customer')

plt.subplot(1, 2, 2)
plt.hist(df_new['M'])
plt.xlabel('M')
plt.ylabel('cnt_customer')
plt.show()
```

```python
# 상자그림을 통한 이상치 확인
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(df_new['F'])
plt.xlabel('F')

plt.subplot(1, 2, 2)
plt.boxplot(df_new['M'])
plt.xlabel('M')
plt.show()
```

## 1.2 F, M feature 기반으로 군집분석 실시, 필요시 이상값 보정

```python
# Invoice 유니크한 값 구하기
df['InvoiceDate'].unique().shape
```

```python
# 2084일까지 살 수 있는데, 5903이 생긴 것은 이상하다.
# 따라서 F 변수에서 2084가 넘는 값 몇 개인지 확인

df_new[df_new['F'] > 2084]['F'].count()
```

```python
# 2개 이므로 삭제한다.

df_new = df_new[df_new['F'] <= 2084]
df_new.describe()
```

```python
# 다시 상자그림 그려보기
plt.figure(figsize = (10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(df_new['F'])
plt.xlabel('F')

plt.subplot(1, 2, 2)
plt.boxplot(df_new['M'])
plt.xlabel('M')
plt.show()
```

### 군집분석 실시 (feat.비계층적 군집분석)

k-means 군집분석

```python
# 군집분석 하기 전에 스케일링
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()

df_new_scaled = pd.DataFrame(scaler.fit_transform(df_new), columns = df_new.columns)
```

```python
# 시각화랑 같이 하기
from sklearn.cluster import KMeans
import seaborn as sns

k = [3, 4, 5, 10]
palette = ['green', 'red', 'blue', 'yellow', 'gray', 'brown', 'pink', 'purple', 'black', 'orange']
for i in k:
    model = KMeans(n_clusters = i, random_state = 731)
    model.fit(df_new_scaled)
    df_new_scaled['cluster'] = model.fit_predict(df_new_scaled)
    sns.scatterplot(x='F', y='M', hue = 'cluster', data = df_new_scaled, palette = palette[:i])
    plt.title('K = %d results'%i)
    plt.show()
```

```python
# 저 위에 두 점을 없애고 다시 해야할 것 같음
df_new_scaled.pop('cluster')
df_new_scaled = df_new_scaled[df_new_scaled['F'] < 8]

k = [3, 4, 5, 10]
palette = ['green', 'red', 'blue', 'yellow', 'gray', 'brown', 'pink', 'purple', 'black', 'orange']
for i in k:
    model = KMeans(n_clusters = i, random_state = 731)
    model.fit(df_new_scaled)
    df_new_scaled['cluster'] = model.fit_predict(df_new_scaled)
    sns.scatterplot(x='F', y='M', hue = 'cluster', data = df_new_scaled, palette = palette[:i])
    plt.title('K = %d results'%i)
    plt.show()
```

## 1.3 군집 결과의 적합성을 군집 내 응집도, 군집 간 분리도의 개념을 사용해서 서술

군집분석의 원리는 크게 두 가지이다.
1. 군집 내 응집도 최대화
2. 군집 내 분리도 최대화

* 군집 내 응집도 최대화는 같은 군집 내 응집도를 최대화하는 것으로, 같은 군집 내 객체들의 거리를 최소화하는 것이다.
* 군집 내 분리도 최대화란 다른 군집 간 거리를 최대화하는 것이다.

이때, inertia의 값이 작을수록 응집도가 높게 군집이 잘 되었다고 생각할 수 있다.

```python
inertia_list = []
for i in range(2, 11):
    model = KMeans(n_clusters = i, random_state = 731)
    model.fit(df_new_scaled)
    inertia_list.append(model.inertia_)

print(inertia_list)
```

이것을 시각화해보면 엘보우기법을 통해 다음과 같이 나타낼 수 있다.

```python
# 엘보우 메소드를 이용하여 확인해보기
# scree plot 느낌(주성분 분석)

from yellowbrick.cluster import KElbowVisualizer

df_new_scaled.pop('cluster')

model_test = KMeans()
visualizer = KElbowVisualizer(model_test, k=(1, 10))
visualizer.fit(df_new_scaled)
```

## 3-2 연매출이 3000, 4000, 5000이었다면 연평균 몇배가 증가한 것인가?

증가율의 평균을 구하기 위해 기하평균을 사용한다.

```python
r1 = ((4000-3000)/3000) + 1
r2 = ((5000-4000)/4000) + 1

import numpy as np
g_mean = np.sqrt(r1*r2)
g_mean
```

따라서, 연평균 29% 정도 증가했다고 판단할 수 있다.

## 3-4 표본 10개의 분산이 90일 때 신뢰도 95%로 모분산의 신뢰구간을 추정

신뢰도 95%에 대한 신뢰구간을 구하는 것이므로 $\alpha$의 값은 각각 0.025, 0.975이다.
이때, 자유도는 10-1=9이므로 카이제곱 통계량은 각각 2.70, 19.02이다.

```python
n = 10
s_squared = 90
chi_1 = 2.70
chi_2 = 19.02
```

$\dfrac{(n-1)s^{2}}{{\chi^{2}}_{\frac{\alpha}{2}}} \le \sigma^{2} \le \dfrac{(n-1)s^{2}}{{\chi^{2}}_{1-\frac{\alpha}{2}}}$

위의 공식을 이용하여 다음과 같이 신뢰구간을 구할 수 있다.

```python
lower_bound = ((n-1)*s_squared) / (chi_2)
upper_bound = ((n-1)*s_squared) / (chi_1)

print(lower_bound, upper_bound)
```

따라서 95%의 신뢰도로 모분산은 42.5867 이상, 300 이하라고 추정할 수 있다.

# 6번 1개년 50억원, 2개년 60억원, 3개년 70억원의 예산을 가지고 NPV(순현재가치)가 가장 높아지는 안을 제시하시오.

![image](https://github.com/Datamanim/datarepo/blob/main/adp/25/problem6.png?raw=true)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(data = {"1개년" : [1.1, 1.15, 1.12, 1.13, 1.16],
                         "2개년" : [1.2, 1.14, 1.11, 1.25, 1.30],
                         "3개년" : [1.15, 1.19, 1.3, 1.2, 1.24]},
                 index = ['1안', '2안', '3안', '4안', '5안'])
df_array = df.to_numpy()
# print(df_array)

budget = np.array([50, 60, 70])
# print(budget)

NPV = df_array.dot(budget)
print(NPV)
```

따라서 순현재가치(NPV)가 가장 높아지는 안은 5안이다.
