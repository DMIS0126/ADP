# 최근접 이웃(KNN)

* 유사한 특성을 가진 데이터는 유사한 범주에 속하는 경향이 있다는 가정으로 데이터를 분류하는 머신러닝 기법이다.
* KNN의 기본 원리
  * 학습데이터를 그대로 저장한 뒤 새로운 데이터 포인트에 대해 학습데이터에서 가장 가까운 k개의 데이터 포인트를 찾아 그것들로부터 새로운 데이터포인트의 범주를 라벨링
* 단순한 방식이지만 손글씨, 위성 이미지 분석 등 다양한 분류 및 회귀 문제에서 높은 예측력을 보인다.
* 또한, 비모수적인 방식을 사용하기 때문에 결정경계가 매우 불규칙한 분류 상황에서 종종 높은 예측성능을 보인다.

### KNN 분류
* KNN은 유클리드 거리를 많이 사용한다.
  * 그런데 KNeighborsClassifies 함수에서 metirc의 default는 minkowski이다.
* 추가 내용은 이어서 쓸 예정

### KNN 회귀
* 모든 시점에서 예측된 값들을 하나의 선으로 연결하면 예측에 사용된 데이터들에 대한 표준 데이터를 표시할 수 있으므로 시계열분석에서도 종종 사용할 수 있다.

```python
# KNeighborsClassifier
import pandas as pd

liver = pd.read_csv('./data/Indian_Liver_Patient.csv', names = ['Age',
                               'Gender',
                               'Total_Bilirubin',
                               'Direct_Bilirubin', 'Alkphos Alkaline Phosphotase',
                               'Alamine_Aminotransferase',
                               'Aspartate_Aminotransferase',
                               'Total_Protiens',
                               'Albumin',
                               'Albumin_and_Globulin_Ratio',
                               'Dataset'])

print(liver.info())

# Gender 변수를 one-hot Encoding하기
liver.loc[:,'Gender'] = liver.loc[:, 'Gender'].map({'Female' : 0, 'Male' : 1})
liver['Gender'] = liver['Gender'].astype('int')
print(liver.dtypes, '\n=====\n', liver['Gender'].head())

# Ratio_Albumin_and_Globulin_Ratio 컬럼에 4개의 결측치가 있고, 이는 전체 데이터에 비해 작은 크기이므로 삭제
liver.dropna(axis = 0, inplace = True) # axis = 0(결측치가 있는 행 삭제)(default), 1(결측치가 있는 열 삭제)
print(liver.info())

# 설명변수와 타깃변수(Dataset) 설정한 뒤 7:3 비율로 학습데이터와 평가데이터로 분리하기
y=liver.pop('Dataset')
x=liver

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1, stratify=y)
# stratify : None이 아닌 경우 데이터는 지정한 변수를 기준으로 계층화되어 해당 변수의 비율을 유지하도록 추출
    # 즉, 층화임의추출 여부를 결정한다.

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# KNeighborsClassifier를 이용하여 KNN분류기 생성하기
from sklearn.neighbors import KNeighborsClassifier

KNC = KNeighborsClassifier(n_neighbors=15, weights='uniform')
KNC.fit(x_train, y_train)

# 모델의 성능지표 확인하기
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

pred = KNC.predict(x_test)

test_cm = confusion_matrix(y_test, pred)
test_acc = accuracy_score(y_test, pred)
test_prc = precision_score(y_test, pred)
test_rcll = recall_score(y_test, pred)
test_f1 = f1_score(y_test, pred)

print('혼동행렬\n', test_cm)
print('정확도\t{}%'.format(round(100*test_acc, 2)))
print('정밀도\t{}%'.format(round(100*test_prc, 2)))
print('재현율\t{}%'.format(round(100*test_rcll, 2)))
print('f1_score\t{}%'.format(round(100*test_f1, 2)))
```

```python
# KNeighborsRegressor
# KNeighborsClassifier 함수와 동일한 파라미터를 사용한다.

# 임의의 데이터로 KNN을 사용한 회귀분석 해보기
import numpy as np

np.random.seed(0)
x = np.sort(5*np.random.rand(400, 1), axis=0)
T= np.linspace(0, 5, 500)[:, np.newaxis]
y = np.sin(x).ravel()

print(x[:10])
print(T[:10])
print(y[:10])

# 타깃데이터(y)에 노이즈 추가하기
y[::1] += 1 * (0.5 - np.random.rand(400))

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

# KNeighborsRegressor를 이용하여 weights를 다르게 적용한 KNN 추정기 생성하기
from sklearn.neighbors import KNeighborsRegressor
knn_uni = KNeighborsRegressor(weights='uniform', n_neighbors=20)
knn_dis = KNeighborsRegressor(weights='distance', n_neighbors=20)

knn_uni.fit(x_train, y_train)
knn_dis.fit(x_train, y_train)

uni_pred = knn_uni.predict(x_test)
dis_pred = knn_dis.predict(x_test)

from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

preds = [uni_pred, dis_pred]
weights = ['uniform', 'distacne']
evls = ['mse', 'rmse', 'mae']

results = pd.DataFrame(index = weights, columns=evls)

for pred, nm in zip(preds, weights):
   mse = mean_squared_error(y_test, pred)
   mae = mean_absolute_error(y_test, pred)
   rmse = np.sqrt(mse)

   results.loc[nm]['mse'] = round(mse, 2)
   results.loc[nm]['rmse'] = round(rmse, 2)
   results.loc[nm]['mae'] = round(mae, 2)

# uniform과 distance의 차이를 직관적으로 비교하기 위해 시각화하기
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 5))

for i, weights in enumerate(['uniform', 'distance']):
    knn = KNeighborsRegressor(n_neighbors=20, weights=weights)

    y_ = knn.fit(x, y).predict(T)

    plt.subplot(2, 1, i+1)
    plt.scatter(x, y, color='darkorange', label='data')
    plt.plot(T, y_, color='navy', label='prediction')
    plt.axis('tight')
    plt.legend(loc = 'upper right')
    plt.title("KNeighborsRegressor (k=%i, weight='%s')" % (5, weights))

plt.tight_layout()
plt.show()
```
