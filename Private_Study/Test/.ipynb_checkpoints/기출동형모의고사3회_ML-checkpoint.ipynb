5. 샘플링 기법을 설명하고 비교한 뒤 2개 기법을 설명하고 근거를 제시하시오.

#### * 오버 샘플링 기법은 비중이 적은 데이터를 추가로 생성해 수를 늘려 데이터 불균형을 극복하는 기법이다.
언더 샘플링은 데이터 손실의 문제로 예측 성능이 저하되는 단점이 있으므로, 일반적으로는 불균형한 데이터를 처리하는 방식으로 오버 샘플링을 사용한다.

#### * 이때, 오버 샘플링은 Random Oversampling과 SMOTE로 구분할 수 있는데,
Random Oversampling은
* 소수 클래스에 속하는 데이터의 관측치를 복사하는 방식으로 데이터를 증식한다.
* 데이터를 단순 복사하는 방식이므로 기존의 데이터와 동일한 복제 데이터를 생성한다.
* 이에 따라 소수 클래스에 과적합이 발생할 가능성이 있다는 단점이 있지만, 사용방법이 간단하다는 장점이 있다.

한편, SMOTE는
* 적은 데이터 세트에 있는 개별 데이터들의 k-최근접 이웃을 찾아, 해당 데이터와 k개 이웃들의 차이를 일정한 값으로 만들어 기존 데이터와 약간의 차이를 지닌 새로운 데이터를 생성하는 방식이다.
* SMOTE는 Random Oversampling보다 처리 속도가 느리다는 단점이 잇지만, 데이터를 단순히 동일하게 증식시키는 방식이 아니기 때문에, 과적합 문제를 예방할 수 있는 장점이 있다.

이에 따라 Random Oversampling과 SMOTE를 선정한다.

```python
# 우선, object 형태의 변수인 deposit_type을 LabelEncoding한다.
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
hotel['is_canceled'] = hotel['is_canceled'].astype('int')
hotel['is_repeated_guest'] = hotel['is_repeated_guest'].astype('int')

hotel['deposit_type'] = encoder.fit_transform(hotel['deposit_type'])
y=hotel['is_canceled']
x=hotel[hotel.columns.difference(['is_canceled'])]
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
ros = RandomOverSampler(random_state = 731)
smt = SMOTE(random_state = 731)

import time
# Random Oversampling
ros_start_time = time.time() # 시작 시간
x_ros, y_ros = ros.fit_resample(x, y)
ros_end_time = time.time() # 종료 시간
print('ros_time : ', ros_end_time - ros_start_time)

# SMOTE
smt_start_time = time.time() # 시작 시간
x_smt, y_smt = smt.fit_resample(x, y)
smt_end_time = time.time() # 종료 시간
print('smt_time : ', smt_end_time - smt_start_time)
```

7. 오버 샘플링 데이터와 원데이터를 사용해 정확도 측면 모델 하나와 속도 측면의 모델 하나를 선정하고, 그 이유를 설명하시오.

```python
# 정확도 : xgboost
# 속도 : Logistic

# 그 전에 robust scaling 진행
from sklearn.preprocessing import RobustScaler
rb_scaler = RobustScaler()
x = pd.DataFrame(rb_scaler.fit_transform(x), columns = x.columns)
x_ros = pd.DataFrame(rb_scaler.fit_transform(x_ros), columns = x_ros.columns)
x_smt = pd.DataFrame(rb_scaler.fit_transform(x_smt), columns = x_smt.columns)
```

```python
# 정확도 : xgboost
from xgboost import XGBClassifier
xgb = XGBClassifier(n_estimators = 300, random_state = 731)
from sklearn.metrics import classification_report

# 1. 원데이터
print('원데이터')
xgb_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=731, stratify=y)
xgb.fit(x_train, y_train)
print('train 정확도 : ', xgb.score(x_train, y_train), '\n')
xgb_pred = xgb.predict(x_test)
print(classification_report(y_test, xgb_pred))
xgb_end_time = time.time()
print('time : ', xgb_end_time - xgb_start_time)
print('='*20)

# 2. ros 데이터
print('ros 데이터')
xgb_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.3, random_state=731, stratify=y_ros)
xgb.fit(x_train, y_train)
print('train 정확도 : ', xgb.score(x_train, y_train), '\n')
xgb_pred = xgb.predict(x_test)
print(classification_report(y_test, xgb_pred))
xgb_end_time = time.time()
print('time : ', xgb_end_time - xgb_start_time)

# 3. smt 데이터
print('smt 데이터')
xgb_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_smt, y_smt, test_size=0.3, random_state=731, stratify=y_smt)
xgb.fit(x_train, y_train)
print('train 정확도 : ', xgb.score(x_train, y_train), '\n')
xgb_pred = xgb.predict(x_test)
print(classification_report(y_test, xgb_pred))
xgb_end_time = time.time()
print('time : ', xgb_end_time - xgb_start_time)
```

```python
# 속도 : Logistic
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 731, n_estimators=100, max_depth=100, criterion='entropy')

# 1. 원데이터
print('원데이터')
rf_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=731, stratify=y)
rf.fit(x_train, y_train)
print('train 정확도 : ', rf.score(x_train, y_train), '\n')
rf_pred = rf.predict(x_test)
print(classification_report(y_test, rf_pred))
rf_end_time = time.time()
print('time : ', rf_end_time - rf_start_time)
print('='*20)

# 2. ros 데이터
print('ros 데이터')
rf_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.3, random_state=731, stratify=y_ros)
rf.fit(x_train, y_train)
print('train 정확도 : ', rf.score(x_train, y_train), '\n')
rf_pred = rf.predict(x_test)
print(classification_report(y_test, rf_pred))
rf_end_time = time.time()
print('time : ', rf_end_time - rf_start_time)

# 3. smt 데이터
print('smt 데이터')
rf_start_time = time.time()
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_smt, y_smt, test_size=0.3, random_state=731, stratify=y_smt)
rf.fit(x_train, y_train)
print('train 정확도 : ', rf.score(x_train, y_train), '\n')
rf_pred = rf.predict(x_test)
print(classification_report(y_test, rf_pred))
rf_end_time = time.time()
print('time : ', rf_end_time - rf_start_time)
```

#### * 해보니까 RF가 짱이다. xgboost보다 좋네
#### * 그리고 ros가 smt보다 좋은 성능을 나타내었다.
#### * accuracy를 보면, 원데이터에 비해 정확도가 높아졌음을 알 수 있다.
#### * macro_avg를 보면, 모두 train 정확도보다 낮고, 차이가 많이 나므로 데이터 학습 시 과적합이 발생했음을 알 수 있다. 그래도 ros일 때의 train 정확도와 macro_avg 차이가 가장 작다.
